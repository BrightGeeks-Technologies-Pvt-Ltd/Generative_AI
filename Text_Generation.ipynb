```python
# Import required libraries from transformers and torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch
```

```python
# Load the GPT-2 model and tokenizer
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')

# Check if GPU is available and move the model to GPU if it is, otherwise use CPU
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model = model.to(device)
```

```python
# Define a text generation function that resolves the warnings
def generate_text_gpt2(prompt, max_length=100, temperature=1.0, do_sample=True):
    # Tokenize the input prompt and create the attention mask
    input_ids = tokenizer.encode(prompt, return_tensors='pt').to(device)
    attention_mask = torch.ones(input_ids.shape, device=device)  # Ensuring a valid attention mask

    # Generate text using the GPT-2 model
    output = model.generate(
        input_ids,
        attention_mask=attention_mask,  # Explicitly pass attention mask
        max_length=max_length,
        temperature=temperature,  # Use temperature for sampling
        do_sample=do_sample,  # Enable sampling mode
        num_return_sequences=1,   # Number of sequences to return
        pad_token_id=tokenizer.eos_token_id  # Avoid padding-related issues
    )

    # Decode the generated tokens back into text
    generated_text = tokenizer.decode(output[0], skip_special_tokens=True)

    return generated_text
```

The prompt


```python
# Example prompt for the demonstration
prompt = "Artificial Intelligence is important because"

# Generate text
generated_text = generate_text_gpt2(prompt, max_length=100, temperature=0.7, do_sample=True)

# Print the generated text
print("Generated Text:")
print(generated_text)
```
